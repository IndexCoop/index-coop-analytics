## How does INDEX token price move with ETH price? 

#### We'll visualize how the prices have each changed since INDEX token genesis and then overlay scaled token prices relative to their peak price. 

*** INDEX tokens were first tracked by Coingecko (where the data is pulled from) on October 6, 2020. 

#### Spearman's correlation of INDEX price to ETH price since genesis: r = 0.878, 95% Confidence Interval [0.85, 9]
#### Spearman's correlation of INDEX price to ETH price in the last 90 days: r = 0.871, 95% Confidence Interval [0.81, 0.91]

The correlation between INDEX and ETH price has been pretty high which means that INDEX prices move extremely similarly that that of ETH. This correlation appears to hold in the past 90 days.

### Another view of the INDEX / ETH price relationship

#### There's a strong relationship between ETH and INDEX price; 95% confidence intervals of the relationship are shown in the light blue ranges. A linear regression model for this would be INDEX(USD price) = 0.015*ETH(USD price) - 3.474(USD)

## How much variation is there in INDEX price?

#### Since genesis, standard deviation of INDEX price has been $15.27 with 95% confidence interval (14.15,16.44). Measured in ETH, standard deviation of INDEX price has been 0.004698 ETH 95%CI (0, 0.01). I suspect there is a lower bound to reporting the bottom end of the 95% CI range.
